```python
import pandas as pd

df = pd.read_excel("data/sales_raw.xlsx")
df.head()
```

```python
# Check column info & null values
df.info()

# Summary statistics
df.describe()

# Check for missing values
df.isnull().sum()
```

```python
# Total Sales per Region
sales_region = df.groupby("Region")["Total_Sales"].sum()
print(sales_region)
```

```python
# Total Profit per Product Category
profit_category = df.groupby("Product_Category")["Profit"].sum()
print(profit_category)
```

```python
top_products = df.groupby("Product_Name")["Quantity"].sum().sort_values(ascending=False)
print(top_products.head(10))  # Top 10 products
```

```python
import matplotlib.pyplot as plt

# Sales per Region
sales_region.plot(kind="bar", title="Total Sales per Region")
plt.ylabel("Total Sales")
plt.show()

# Profit per Category
profit_category.plot(kind="bar", title="Total Profit per Category", color='green')
plt.ylabel("Profit")
plt.show()

# Top 10 Products
top_products.head(10).plot(kind="bar", title="Top 10 Selling Products", color='orange')
plt.ylabel("Quantity Sold")
plt.show()
```
